```python
import csv
from functools import reduce
import json
```

```python

data = []
with open('/content/911_Calls_for_Service_(Last_30_Days).csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        data.append(row)
```

```python
filtered_data = list(filter(lambda x: x['zip_code'] != '' and x['neighborhood'] != '', data))
```

```python
numeric_data = [
    {
        'total_response_time': float(row['totalresponsetime']) if row['totalresponsetime'] != '' else 0,
        'dispatch_time': float(row['dispatchtime']) if row['dispatchtime'] != '' else 0,
        'total_time': float(row['totaltime']) if row['totaltime'] != '' else 0
    }
    for row in filtered_data
]
```

```python
total_response_time = reduce(lambda x, y: x + y['total_response_time'], numeric_data, 0)
total_dispatch_time = reduce(lambda x, y: x + y['dispatch_time'], numeric_data, 0)
total_total_time = reduce(lambda x, y: x + y['total_time'], numeric_data, 0)
```

```python
average_response_time = total_response_time / len(numeric_data) if len(numeric_data) > 0 else 0
average_dispatch_time = total_dispatch_time / len(numeric_data) if len(numeric_data) > 0 else 0
average_total_time = total_total_time / len(numeric_data) if len(numeric_data) > 0 else 0
```

```python
 print("Total rata-rata waktu respons:", average_response_time)
 print("Total rata-rata waktu pengiriman:",average_dispatch_time)
 print("Total rata-rata waktu keseluruhan:", average_total_time)
```

```python
neighborhood_data = {}
for row in filtered_data:
    if row['neighborhood'] not in neighborhood_data:
        neighborhood_data[row['neighborhood']] = []
    neighborhood_data[row['neighborhood']].append(row)
```

```python
# Step 7: Calculate stats for each neighborhood
neighborhood_stats = []
for neighborhood, incidents in neighborhood_data.items():
    numeric_data = [
        {
            'total_response_time': float(row['totalresponsetime']) if row['totalresponsetime'] != '' else 0,
            'dispatch_time': float(row['dispatchtime']) if row['dispatchtime'] != '' else 0,
            'total_time': float(row['totaltime']) if row['totaltime'] != '' else 0
        }
        for row in incidents
    ]
    total_response_time = reduce(lambda x, y: x + y['total_response_time'], numeric_data, 0)
    total_dispatch_time = reduce(lambda x, y: x + y['dispatch_time'], numeric_data, 0)
    total_total_time = reduce(lambda x, y: x + y['total_time'], numeric_data, 0)
    average_response_time = total_response_time / len(numeric_data) if len(numeric_data) > 0 else 0
    average_dispatch_time = total_dispatch_time / len(numeric_data) if len(numeric_data) > 0 else 0
    average_total_time = total_total_time / len(numeric_data) if len(numeric_data) > 0 else 0
    neighborhood_stats.append({
        'neighborhood': neighborhood,
        'average_response_time': average_response_time,
        'average_dispatch_time': average_dispatch_time,
        'average_total_time': average_total_time
    })

# Print neighborhood stats
for stats in neighborhood_stats:
    print("Lingkungan:", stats['neighborhood'])
    print("Rata-rata waktu respon:", stats['average_response_time'])
    print("Rata-rata waktu pengiriman:", stats['average_dispatch_time'])
    print("Total waktu dalam rata-rata:", stats['average_total_time'])
```

```python
import json

output_file = 'neighborhood_stats.json'

with open(output_file, 'w') as json_file:
    json.dump(neighborhood_stats, json_file, indent=4)

print(f"File JSON '{output_file}' telah berhasil dibuat.")
```
